## Creating dummy variables

```python
reason_dummies = pd.get_dummies(df['Reason for Absence'])
```

```python
reason_dummies
```

### Checking if there is only one reason for each row

```python
reason_dummies['check'] = reason_dummies.sum(axis=1)
reason_dummies.head()
```

```python
reason_dummies['check'].sum(axis=0) # sum on vertical
```

```python
reason_dummies['check'].unique()
```

We have determined that there is only one reason the employee was away from work, as each row has a '1' at their check. Thus, we can now drop the check column, as it will interfere.

```python
reason_dummies.drop(['check'], axis=1)
```

We need to remove reason 0 to avoid multicollinearity issues -> have n-1 dummy variables.

```python
reason_dummies = pd.get_dummies(df['Reason for Absence'], drop_first=True)
```

## Combining dummies with df

```python
df.drop(['Reason for Absence'], axis=1, inplace=True)
```

We will separate the reasons for absence in four groups. Group 1 relating to disease, Group 2 relating to pregnancy, Group 3 relating to poison, Group 4 relating to more "light reasons" for absence.

```python
reason_t1 = reason_dummies.loc[:, 1:14].max(axis=1)
reason_t2 = reason_dummies.loc[:, 15:17].max(axis=1)
reason_t3 = reason_dummies.loc[:, 18:21].max(axis=1)
reason_t4 = reason_dummies.loc[:, 22:28].max(axis=1)
```

```python
df = pd.concat([df, reason_t1, reason_t2, reason_t3, reason_t4], axis=1)
df
```

# Working with Dates

```python
type(df['Date'][0])
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y') # timestamp
```

```python
df.head()
```

## Extracting the month

```python
lst_months = []

for i in range(df.shape[0]):
    lst_months.append(df['Date'][i].month)
```

```python
df['Month'] = lst_months
df.head()
```

## Extracting day of week

```python
df['Day of Week'] = df['Date'].apply(lambda x: x.weekday())
```

```python
df.drop(['Date'], axis=1, inplace=True)
```
